**Geometry optimization and  charges :**
Before proceeding to analyses, such as adsorption, a system should be in its most stable and lowest-energy configuration. To insure that, geometry optimization is used, minimizing the system's total energy.
To predict chemical reactivity, bonding and the properties of materials, the distribution of electron charges in a system is calculated.

**Pore analysis :**
The structure of porous materials can be determined on the basis of pore size, volume and distribution. Its analysis is essential for understanding how asborbates interact with the internal surfaces of materials.

**Isotherm calculations :**
To predict the absorption capacity and assess the efficiency of a material in storing and separating gases, isotherm calculations are essential. They describe how the quantity of gas absorbed by a material varies as a function of pressure at a constant temperature.

```python
!pip install pyiast
import numpy as np
import matplotlib.pyplot as plt
import pyiast
```

```python
df_ch4 = pd.read_csv("CH4_IRMOF1.csv")

ch4_isotherm = pyiast.InterpolatorIsotherm(df_ch4,
                                    loading_key="Loading(mmol/g)",
                                    pressure_key="Pressure(bar)",
                                    fill_value=df_ch4['Loading(mmol/g)'].max())
```

```python
df_co2 = pd.read_csv("CO2_IRMOF1.csv")

co2_isotherm = pyiast.InterpolatorIsotherm(df_co2,
                                    loading_key="Loading(mmol/g)",
                                    pressure_key="Pressure(bar)",
                                    fill_value=df_ch4['Loading(mmol/g)'].max()) 
```

```python
!pip install pyiast
import numpy as np
import matplotlib.pyplot as plt
import pyiast

pressure_values = np.arange(0.1, 4.1, 0.2) 
q_values = []

for total_pressure in pressure_values:
    y = np.array([0.4, 0.6])  # gas mole fractions
    
    # Perform IAST calculation
    q = pyiast.iast(total_pressure * y, [co2_isotherm, ch4_isotherm], verboseflag=True)
    q_values.append(q)

q_values = np.array(q_values)

plt.figure(figsize=(10, 6))
plt.plot(pressure_values, q_values, marker='o', color='b', linestyle='-', linewidth=2)

plt.title("IAST Calculation: q vs. Total Pressure", fontsize=16)
plt.xlabel("Total Pressure (bar)", fontsize=14)
plt.ylabel("Adsorption Loading (q)", fontsize=14)
plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
